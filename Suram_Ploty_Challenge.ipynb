```python
# Group-1
import pandas as pd
import plotly.express as px

# Load Pokemon dataset
pokemon = pd.read_csv("C:\\Users\\Harsh\\Interactive_data\\pokedex.csv")


```

```python
# Remove irrelevant columns
pokemon = pokemon.drop(columns=['Hidden Ability',
                                     'EV Worth',
                                     'Gender',
                                     'Egg Group II',
                                    ])

```

```python
# Importing required libraries
import plotly.express as px

# Defining the variables
x = 'Catch'
y = 'Total'
color = 'Type I'
title = 'Pokemon Stats Organized by Type'

# Creating the scatter plot
fig = px.scatter(pokemon, x=x, y=y, color=color, title=title)

# Displaying the scatter plot
fig.show()
```

```python


# Creating our first scatter plot with this data. Observing Pokemon catch rates with their stat totals based on their typings
fig = px.scatter(pokemon, x='Catch', y='Total', color='Type I', title='Pokemon Stats Organized by Type',
                 labels={'Catch': 'Catch Rate', 'Total': 'Stat Total', 'Type I': 'Pokemon Type'},
                 hover_data=['Pokemon'])

# Enhancing this graph by assigning a variable to the size of bubbles, increasing the spacing/size of the graph
fig.update_traces(marker=dict(size=10), selector=dict(mode='markers'))

# Customizing the bubble color
fig.update_traces(showlegend=True)

# Adding hover information
fig.update_layout(hovermode='closest',
                  hoverlabel=dict(bgcolor='white', bordercolor='gray', font_size=12),
                  hoverlabel_align='left')

# Adding axis labels
fig.update_xaxes(title_text='Catch Rate')
fig.update_yaxes(title_text='Stat Total')

fig.show()
```

```python
# Group-2
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np

# Load Water Quality dataset
Water_Quality = pd.read_csv("C:\\Users\\srika\\Interactive_data\\wq.csv")

# Histograms
fig_hist_site = px.histogram(Water_Quality, x="Site_Id", histfunc='count', color='Site_Id',
                             title='Count of Readings from Each Site Id')
fig_avg_do_by_site_year = px.histogram(Water_Quality, x='Year', y='Dissolved Oxygen (mg/L)', color="Site_Id",
                                       barmode='group', histfunc='avg',
                                       facet_col='Site_Id', facet_col_wrap=3, text_auto=True,
                                       height=700, title='Average Dissolved Oxygen by Year for Each Site')


# violin plot of water depth by site ID

# Extract unique Site IDs and corresponding data
Site_Id= Water_Quality['Site_Id'].unique()
fig_violin_depth_site = [Water_Quality[Water_Quality['Site_Id'] == site]['Water Depth (m)'].values for site in Site_Id]

# Create a violin plot
plt.violinplot(fig_violin_depth_site, showmeans=False, showmedians=True)

# Set x-axis ticks and labels
plt.xticks(ticks=np.arange(1, len(Site_Id) + 1), labels=Site_Id)

# Set labels and title
plt.xlabel('Site ID')
plt.ylabel('Water Depth (m)')
plt.title('Violin Plot of Water Depth by Site ID')


fig_hist_site.show()
fig_avg_do_by_site_year.show()


```

```python
# Create a violin plot
violin = plt.violinplot(fig_violin_depth_site, showmeans=False, showmedians=True)

# Set the facecolor of the violin
for pc in violin["bodies"]:
    pc.set_facecolor("blue")

# Set x-axis ticks and labels
plt.xticks(ticks=np.arange(1, len(Site_Id) + 1), labels=Site_Id)

# Set labels and title
plt.xlabel('Site ID')
plt.ylabel('Water Depth (m)')
plt.title('Violin Plot of Water Depth by Site ID')

plt.show()

```

```python


# Create a 2x2 axes
fig, axs = plt.subplots(2, 2)

# Water depth by site ID on the first column, first row
axs[0, 0].violinplot(fig_violin_depth_site, showmeans=False, showmedians=True)
axs[0, 0].set_xticks(np.arange(1, len(Site_Id) + 1))
axs[0, 0].set_xticklabels(Site_Id)
axs[0, 0].set_xlabel('Site ID')
axs[0, 0].set_ylabel('Water Depth (m)')
axs[0, 0].set_title('Violin Plot of Water Depth by Site ID')

# Another violin plot on the first column, second row
axs[1, 0].violinplot(fig_violin_depth_site, showmeans=False, showmedians=True)
axs[1, 0].set_xticks(np.arange(1, len(Site_Id) + 1))
axs[1, 0].set_xticklabels(Site_Id)
axs[1, 0].set_xlabel('Site ID')
axs[1, 0].set_ylabel('Water Depth (m)')
axs[1, 0].set_title('Another Violin Plot')

plt.show()

```

```python
# Group-3
import pandas as pd
import plotly.express as px
import plotly.io as pio

# Load Jobs in Data dataset
Jobs_in_Data = pd.read_csv("C:\\Users\\srika\\Interactive_data\\jobs_in_data.csv")

# Box Plot for Salary Distribution by Experience Level
fig_box_exp_salary = px.box(Jobs_in_Data, x='experience_level', y='salary_in_usd',
                            title='Salary Distribution by Experience Level',
                            labels={'salary_in_usd': 'Salary in USD', 'experience_level': 'Experience Level'},
                            category_orders={"experience_level": ["Entry-level", "Mid-level", "Senior", "Executive"]})

# Convert figure to dictionary
fig_dict = fig_box_exp_salary.to_dict()

# Convert figure to JSON string
fig_json = fig_box_exp_salary.to_json()

# Convert JSON string back to Plotly figure
fig_from_json = pio.from_json(fig_json)

# Show figures
fig_box_exp_salary.show()

# Enhanced Box Plot with Different Colors
fig_box_exp_salary_colored = px.box(Jobs_in_Data, x='experience_level', y='salary_in_usd',
                                     color='experience_level',
                                     title='Salary Distribution by Experience Level (Colored)',
                                     labels={'salary_in_usd': 'Salary in USD', 'experience_level': 'Experience Level'},
                                     category_orders={"experience_level": ["Entry-level", "Mid-level", "Senior", "Executive"]})

# Show the colored figure
fig_box_exp_salary_colored.show()

# Box Plot for Salary Distribution by Job Category
fig_box_job_category_salary = px.box(Jobs_in_Data, x='job_category', y='salary_in_usd',
                                     title='Salary Distribution by Job Category',
                                     labels={'salary_in_usd': 'Salary in USD', 'job_category': 'Job Category'})

# Show the figure for job category distribution
fig_box_job_category_salary.show()

# Proposed Plotly Challenge for the Class:
# Salary Distribution by Job Category
fig_box_job_category_salary_challenge = px.box(Jobs_in_Data, x='job_category', y='salary_in_usd',
                                                title='Salary Distribution by Job Category',
                                                labels={'salary_in_usd': 'Salary in USD', 'job_category': 'Job Category'},
                                                color='job_category')

# Show the figure for the challenge
fig_box_job_category_salary_challenge.show()
```

# Salary Distribution 

A box plot is employed to depict the distribution of wages for each job type, providing valuable insights into both the variability and central tendency of salaries.

The complete salary range, encompassing the variation in pay across all occupational categories, spans from $100,000 to $400,000. It is imperative to acknowledge that individual compensation may diverge based on various factors, including experience, geographical location, and company size.

This visualization serves as a powerful tool for stakeholders to make informed decisions and gain a nuanced understanding of compensation structures for different roles.
