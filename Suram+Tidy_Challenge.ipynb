```python
# Assuming your 'messages' data is stored in a variable called data
df = pd.DataFrame(C_M['messages'])

# Columns to drop
Drops_C = ['share', 'photos', 'gifs', 'users', 'reactions', 'is_unsent', 'is_taken_down', 'bumped_message_metadata']

# Drop specified columns
df_drop = df.drop(columns=Drops_C)

# Display the resulting DataFrame after dropping columns
df_drop
```

```python


# Group the DataFrame by multiple columns and count occurrences
grouped_df = df_drop.groupby(['type', 'sender_name', 'content', 'timestamp_ms']).size().reset_index(name='count')

# Display the resulting DataFrame with grouped counts
grouped_df
```
