## Vizualization that i want to learn about 

- I want to learn  the stem plot, which shows discrete numeric values on a vertical plot.

- We could use this for the age variable in the  data, as age is a numeric value but concentrates around certain numbers ending.

- The stem plot would visualize the distribution of age values, making it easy to identify concentrations around values.
 
- This gives a sense of what ages are most represented through the vertical spines at certain numeric values.

```python

import pandas as pd
import matplotlib.pyplot as plt

# Import the CSV file and convert it to a DataFrame object
df = pd.read_csv("C:\\Users\\Harsh\\interactive_Data\\cces_example_data.csv")


# Create an empty figure with two axes
fig, (ax1, ax2) = plt.subplots(1, 2)

# Set the titles for the axes
ax1.set_title('Age Histogram')
ax2.set_title('Political Identity')

# Create a figure with a bar plot of the 'Evaluation' variable
fig2, ax3 = plt.subplots()
df['Evaluation'].value_counts().plot(kind='bar', ax=ax3)
plt.show()
```

```python
# Plotting scatter Plot 
fig3, ax4 = plt.subplots()
df.plot.scatter(x='ConMediaUse', y='ImmiPhobia', ax=ax4, color='green', alpha=0.5)

# Set labels and title
ax4.set_xlabel('ConMediaUse')
ax4.set_ylabel('ImmiPhobia')
ax4.set_title('ConMediaUse vs ImmiPhobia')

# Show the figure
plt.show()
```
