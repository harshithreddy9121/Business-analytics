```python
#Importing libraries
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display
```

```python
# Loading Cricket Data Set
Crick = pd.read_csv("C:\\Users\\Harsh\\Interactive_data\\cricket.csv")
Crick.head()
```

```python
# Plotting Innings vs Runs
image = px.scatter(Crick, x='Innings', y='Runs', labels={'Innings':'Innings','Runs':'Runs'})
image.update_layout(title='Innings Vs Runs', title_x=0.5, title_y=0.95)
image.show()
```

```python
# Plotting Innings vs Runs with player information on hover
image = px.scatter(Crick, x='Innings', y='Runs', 
                 hover_data=['Player', 'Average', 'Country'],
                 labels={'Innings': 'Innings', 'Runs': 'Runs'})

image.update_layout(title='Innings Vs Runs', title_x=0.5, title_y=0.95)
image.show()
```

```python
# Buttons on the graph
image_11 = px.scatter(Crick, x='Innings', y='Runs', 
                 hover_data=['Player', 'Average', 'Country'],
                 labels={'Innings': 'Innings', 'Runs': 'Runs'})

image_11.update_layout(title='Innings Vs Runs', title_x=0.5, title_y=0.95)

image_11.update_layout(
    updatemenus=[
        dict(
            type = "buttons",
            direction = "left",
            buttons=list([
                dict(
                    args=["type", "scatter"],
                    label="Scatter Plot",
                    method="restyle"
                ),
                dict(
                    args=["type", "bar"],
                    label="Bar Chart",
                    method="restyle"
                )
            ]),
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1,
            yanchor="top"
        ),
    ]
)

# Display charts
image_11.show()
```

```python


fitness=pd.read_csv("C:\\Users\\Harsh\\Interactive_data\CardioGoodFitness.csv")
fitness.head()
```

```python
# slider widget  Income variable
slider = widgets.IntSlider(
    value=fitness['Income'].min(),
    min=fitness['Income'].min(),
    max=fitness['Income'].max(),
    step=1000,
    description='Income:'
)

#slider widget
display(slider)
```

```python
# Creating a dropdown widget 
dropdown = widgets.Dropdown(
    options=fitness.columns,
    description='Select Column:',
    disabled=False
)

# dropdown widget
display(dropdown)
```

```python
# Excluding "Product","Income" 
exclude = ["Product", "Income"]
columns = [col for col in fitness.columns if col not in exclude]
column_dropdown = widgets.Dropdown(
    options=columns,
    description='Select Column:',
    disabled=False
)
display(column_dropdown)
```

```python
#  count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Age', hue='Product', data=fitness)
plt.title('Count Plot of Products Grouped by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Product')
plt.show()
```

```python
# update the chart 
def update_chart(selected_feature):
    plt.figure(figsize=(10, 6))
    sns.countplot(x=selected_feature, hue='Product', data=fitness)
    plt.title(f'Count Plot of Products Grouped by {selected_feature}')
    plt.xlabel(selected_feature)
    plt.ylabel('Count')
    plt.legend(title='Product')
    plt.show()

feature_dropdown = widgets.Dropdown(
    options=fitness.columns,
    description='Select Feature:',
    disabled=False
)

output = widgets.interactive_output(update_chart, {'selected_feature': feature_dropdown})

# Display widgets
widgets.VBox([feature_dropdown, output])
```
